Given an integer array nums where the elements are sorted in ascending order, convert it to a height-balanced binary search tree.

 

Example 1:


Input: nums = [-10,-3,0,5,9]
Output: [0,-3,9,-10,null,5]
Explanation: [0,-10,5,null,-3,null,9] is also accepted:

Example 2:


Input: nums = [1,3]
Output: [3,1]
Explanation: [1,null,3] and [3,1] are both height-balanced BSTs.
 

Constraints:

1 <= nums.length <= 104
-104 <= nums[i] <= 104
nums is sorted in a strictly increasing order.

````python

class Solution:
    def sortedArrayToBST(self, nums):
        def build(left, right):
            # base case: invalid segment
            if left > right:
                return None

            # choose middle index as root of this segment
            mid = (left + right) // 2
            node = TreeNode(nums[mid])

            # recursively build left and right subtrees
            node.left = build(left, mid - 1)
            node.right = build(mid + 1, right)

            return node

        return build(0, len(nums) - 1)
````

# Dry run — detailed step-by-step example

Use example: `nums = [-10, -3, 0, 5, 9]`

Call: `Solution().sortedArrayToBST([-10,-3,0,5,9])`

Top-level: `build(0,4)`

* `left = 0`, `right = 4` → `mid = (0+4)//2 = 2` → `nums[2] = 0`

  * Create `node(0)` → this will be the root.

Now build left subtree: `node.left = build(0,1)`

* `build(0,1)` → `mid = (0+1)//2 = 0` → `nums[0] = -10`

  * Create `node(-10)`
  * Left of this: `build(0, -1)` -> base case -> `None`
  * Right of this: `build(1, 1)`:

    * `build(1,1)` → `mid = 1` → `nums[1] = -3` → `node(-3)`

      * left: `build(1,0)` -> `None`
      * right: `build(2,1)` -> `None`
    * So `node(-3)` is leaf.
  * So `node(-10)` has right child `-3`. (Structure for left subtree: root -10, right child -3)

Now build right subtree: `node.right = build(3,4)`

* `build(3,4)` → `mid = (3+4)//2 = 3` → `nums[3] = 5`

  * Create `node(5)`
  * Left of `5`: `build(3,2)` -> `None`
  * Right of `5`: `build(4,4)`:

    * `build(4,4)` → `mid = 4` → `nums[4] = 9` → `node(9)`

      * left: `build(4,3)` -> `None`
      * right: `build(5,4)` -> `None`
    * So `node(9)` is leaf.
  * So `node(5)` has right child `9`.

Final BST structure (visual):

```
       0
      / \
   -10   5
     \     \
     -3     9
```

Recursion stack snapshots (important moments):

* `build(0,4)` → mid 2 → create 0 → calls build(0,1)
* `build(0,1)` → mid 0 → create -10 → calls build(0,-1) -> None then build(1,1)
* `build(1,1)` → create -3 -> returns -> returns up to -10 -> returns to root
* Back to root -> call `build(3,4)` and proceed similarly.

This ensures the tree is height-balanced (the heights of left and right subtrees differ by at most 1 in practice for sorted arrays).

# Time and space complexity

* **Time complexity:** `O(n)`

  * Each element is visited once to create a `TreeNode` and perform constant-time operations. There are `n` nodes and `n` recursive calls overall.

* **Space complexity:** `O(log n)` average (call stack) — `O(n)` worst-case

  * Because the recursion depth equals the height of the tree. For a balanced BST height ≈ `O(log n)` so recursion stack uses `O(log n)` extra space. However, in worst-case (if you used an algorithm that always picked one side), it could degrade to `O(n)`. With the midpoint choice on a sorted array, you get the balanced case and thus `O(log n)` recursion depth. Additionally, the nodes themselves use `O(n)` memory but typically we count extra/auxiliary space and recursion stack.
