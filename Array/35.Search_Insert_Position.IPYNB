Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2
Example 2:

Input: nums = [1,3,5,6], target = 2
Output: 1
Example 3:

Input: nums = [1,3,5,6], target = 7
Output: 4
 

Constraints:

1 <= nums.length <= 104
-104 <= nums[i] <= 104
nums contains distinct values sorted in ascending order.


````python
class Solution:
    def searchInsert(self, nums, target):
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return left  # insert position if not found


# üîç Execution Block
if __name__ == "__main__":
    sol = Solution()
    
    print(sol.searchInsert([1,3,5,6], 5))  # Expected: 2
    print(sol.searchInsert([1,3,5,6], 2))  # Expected: 1
    print(sol.searchInsert([1,3,5,6], 7))  # Expected: 4
    print(sol.searchInsert([1,3,5,6], 0))  # Expected: 0
````

## üß† Strategy: Binary Search

We split the array until we find:

* exact match ‚Üí return index
* first position where `nums[mid] > target` ‚Üí return `mid`
* if target bigger than all ‚Üí return `right + 1`

**Time:** O(log n)
**Space:** O(1)
Corporate-approved optimization. üöÄ
