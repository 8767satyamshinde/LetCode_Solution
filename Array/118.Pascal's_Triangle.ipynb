Given an integer numRows, return the first numRows of Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it as shown:


 

Example 1:

Input: numRows = 5
Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
Example 2:

Input: numRows = 1
Output: [[1]]
 

Constraints:

1 <= numRows <= 30
 

````python
class Solution:
    def generate(self, numRows):
        result = []

        for i in range(numRows):
            # start each row with 1
            row = [1]

            # fill internal values (skip for row 0 and 1)
            if i > 0:
                prev = result[i - 1]  # previous row
                for j in range(1, i):
                    row.append(prev[j - 1] + prev[j])  # sum of two above

                row.append(1)  # end each row with 1

            result.append(row)

        return result
````

# üß† **Line-by-Line Explanation**

Let‚Äôs break it down like a pro:

### `result = []`

* This will store all rows of Pascal‚Äôs Triangle.

### `for i in range(numRows):`

* Loop for each row from `0` to `numRows-1`.

### `row = [1]`

* Every row **starts with 1**.
* Example: `[1]`, `[1, *, 1]`, ‚Ä¶

### `if i > 0:`

* Rows after the first one need internal values.

### `prev = result[i - 1]`

* Fetch the previous row to compute values.

### `for j in range(1, i):`

* Internal elements start at index 1 and end at index `i-1`.

### `row.append(prev[j - 1] + prev[j])`

* Classic Pascal formula:
  `current = above_left + above_right`

### `row.append(1)`

* Every row ends with a **1**.

### `result.append(row)`

* Append the newly formed row.

### `return result`

* Boom ‚Äî Pascal Triangle ready.
