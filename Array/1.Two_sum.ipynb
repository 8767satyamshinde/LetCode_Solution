Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]
 

Constraints:

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.
 

Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

````python
class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
````

````python
sol = Solution()
print(sol.twoSum([2,7,11,15], 9))
print(sol.twoSum([3,2,4], 6))
print(sol.twoSum([3,3], 6))
````

## üîπ Example 1

`nums = [2, 7, 11, 15]`, `target = 9`

### Outer loop: `i = 0`

* `nums[i] = nums[0] = 2`

#### Inner loop: `j = 1`

* `nums[j] = nums[1] = 7`
* Check: `nums[i] + nums[j] = 2 + 7 = 9`
* Compare with target ‚Üí `9 == 9` ‚úÖ **Condition true**
* So: `return [i, j] = [0, 1]`
* Function **stops here**. No more loops.

So for this input, only **one pair** is checked:

* Pair checked: `(i=0, j=1) ‚Üí (2,7)` ‚Üí ‚úÖ matches ‚Üí returns `[0,1]`

## üîπ Example 2

`nums = [3, 2, 4]`, `target = 6`

We go **pair by pair**.

### Outer loop: `i = 0`

* `nums[i] = nums[0] = 3`

#### Inner loop: `j = 1`

* `nums[j] = nums[1] = 2`
* Sum: `3 + 2 = 5`
* Compare: `5 == 6` ‚ùå ‚Üí **no return**, continue

#### Inner loop: `j = 2`

* `nums[j] = nums[2] = 4`
* Sum: `3 + 4 = 7`
* Compare: `7 == 6` ‚ùå ‚Üí still no match

üëâ End of inner loop for `i = 0`.
No pair with `nums[0]` worked.

---

### Outer loop: `i = 1`

* `nums[i] = nums[1] = 2`

#### Inner loop: `j = 2`

* `nums[j] = nums[2] = 4`
* Sum: `2 + 4 = 6`
* Compare: `6 == 6` ‚úÖ **match found**

So we return:

```python
return [1, 2]
```

üìå Pairs checked in order:

1. `(i=0, j=1) ‚Üí (3,2) ‚Üí 5 ‚ùå`
2. `(i=0, j=2) ‚Üí (3,4) ‚Üí 7 ‚ùå`
3. `(i=1, j=2) ‚Üí (2,4) ‚Üí 6 ‚úÖ ‚Üí return [1,2]`

## üîπ Example 3

`nums = [3, 3]`, `target = 6`

### Outer loop: `i = 0`

* `nums[i] = nums[0] = 3`

#### Inner loop: `j = 1`

* `nums[j] = nums[1] = 3`
* Sum: `3 + 3 = 6`
* Compare: `6 == 6` ‚úÖ

Immediately:

```python
return [0, 1]
```

Only one pair exists here, so it‚Äôs super fast.

## üß† What the Loops Are Really Doing (Concept)

* `i` chooses the **first element** of the pair.
* `j` chooses the **second element**, always **after** `i` (`j = i + 1`) so:

  * You **don‚Äôt repeat pairs** (like (0,1) and then (1,0))
  * You **don‚Äôt use the same element twice** (no (i,i))

The algorithm basically says:

> ‚ÄúTry every (i, j) pair where j is ahead of i.
> As soon as one pair sums to target ‚Üí return their indices.‚Äù

That‚Äôs why time complexity is **O(n¬≤)** ‚Üí nested loops, all combinations.
